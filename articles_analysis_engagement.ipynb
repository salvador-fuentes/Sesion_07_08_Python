```python
# Importing packages from our virtual enviroment:
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading file 

We will use pandas pandas to read our csv file 

```python
FILENAME = 'datasets/articles_data_homeproject.csv'

# Index_col hace referencia a que la columna cero sea el índice 
data = pd.read_csv(FILENAME, index_col=0)
data.head()
```

## Checar los tipos de datos que tenemos 

Check datatypes, describe columns and review missing values

```python
# Function para conocer cuántos null tenemos en nuestro dataframe: is.null 
data.isnull().sum().to_frame()
```

```python
columns = [
    'source_name',
    'author',
    'engagement_reaction_count',
    'engagement_comment_count',
    'engagement_share_count',
    'engagement_comment_plugin_count'
]

#Para eliminar los NA de las columnas
data = data[columns].dropna(axis=0, how='any')
```

```python
# Agrupación por source name del engagement reaction count
engagement_reaction_count = data.groupby('source_name')['engagement_reaction_count'].sum()
engagement_reaction_count.to_frame()
```

```python
# Para conocer que fuente de información tiene mayores artículos

highest_source = data['source_name'].value_counts()
highest_source.to_frame()
```

## Data visualization

```python
# Creamos una lista sobre el dataframe de arriba. El de los mayores artículos:
labels = list(highest_source.index.values)
# Creamos una gráfica de pasteles para visualizar el canal de noticias con mayores artículos:
plt.pie(highest_source, labels=labels, autopct= '%1.lf%%')
# Para establecer el título:
plt.title('Highest amount of news')
plt.show()
```
